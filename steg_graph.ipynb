```python
import matplotlib.pyplot as plt
import numpy as np

BPP = [0.0043,4.804444]
PSNR = [16.7153,39.288220]
SSIM = [0.6656,0.952927]
Accuracy = [0.5007,0.943571]
Loss = [0.6931,0.136138]

algorithms = ['Basic Model', 'Enhanced Model']
bar_width = 0.10  # Adjust the bar width here
index = np.arange(len(BPP))
plt.figure(figsize=(10,7))  # Adjust the figsize here
bars = plt.bar(index, BPP, width=bar_width, label='Bits per pixel(Bpp)')
plt.bar(index + bar_width,PSNR, width=bar_width, label='Peak Signal Noise Ratio(Psnr)')
plt.bar(index + 2 * bar_width, SSIM, width=bar_width, label='Structural similarity index of an image(Ssim)')
plt.bar(index + 3 * bar_width, Accuracy, width=bar_width, label='Accuracy')
plt.bar(index + 4 * bar_width,Loss, width=bar_width, label='Loss')

plt.xticks(index + 3 * bar_width, algorithms)  # Adjust the x-axis position
plt.xlabel('Algorithm Name')
plt.ylabel('Evaluation Metrics')
plt.title('Comparison of Basic  and  Enhanced Model Metrics ')
plt.legend()

for i in range(len(algorithms)):
    plt.text(index[i], BPP[i] + 0.3, f'{BPP[i]:.2f}', ha='center')
    plt.text(index[i] + bar_width, PSNR[i] + 0.3, f'{PSNR[i]:.2f}', ha='center')
    plt.text(index[i] + 2 * bar_width, SSIM[i] + 0.3, f'{SSIM[i]:.2f}', ha='center')
    plt.text(index[i] + 3 * bar_width, Accuracy[i] + 0.3, f'{Accuracy[i]:.2f}', ha='center')
    plt.text(index[i] + 4 * bar_width, Loss[i] + 0.3, f'{Loss[i]:.2f}', ha='center')

plt.ylim(0,43)
plt.show()
```
